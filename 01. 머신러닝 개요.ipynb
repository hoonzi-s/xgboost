#**1.3** 회귀 모델 만들기

```python
head = 'https://raw.githubusercontent.com/rickiepark/handson-gb/main/Chapter'
chap = '01'
filename = '/bike_rentals_cleaned.csv'
link = head + chap + filename
df_bikes = pd.read_csv(link)
```

```python
X = df_bikes.iloc[:, :-1]
y = df_bikes.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = tts(X, y, random_state = 2)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error as MSE
```

```python
mse = MSE(y_test, y_pred)
rmse = np.sqrt(mse)
print(f'RMSE: {rmse:0.2f}')
```

```python
df_bikes['cnt'].describe()
```

#**XGBoost**

```python
from xgboost import XGBRegressor
```

```python
xg_reg = XGBRegressor()
xg_reg.fit(X_train, y_train)
y_pred = xg_reg.predict(X_test)
mse = MSE(y_test, y_pred)
rmse = np.sqrt(mse)
print(f'RMSE: {rmse:0.2f}')
```

```python
from sklearn.model_selection import cross_val_score
```

```python
model = LinearRegression()
scores = cross_val_score(model, X, y, scoring = 'neg_mean_squared_error', cv = 10)
```

```python
rmse = np.sqrt(-scores)
print('회귀 RMSE: ', np.round(rmse, 2))
print('RMSE 평균: ', np.round(rmse.mean(), 2))
```

```python
model = XGBRegressor()
scores = cross_val_score(model, X, y, scoring = 'neg_mean_squared_error', cv = 10)
rmse = np.sqrt(-scores)
print('회귀 RMSE: ', np.round(rmse, 2))
print('RMSE 평균: ', np.round(rmse.mean(), 2))
```

#**1.4** 분류 모델 만들기

```python
df_census = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header = None)
df_census
```

```python
df_census.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation',
                  'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 
                   'income']
df_census.head()
```

```python
df_census.info()
```

```python
df_census.isna().sum()
```

```python
df_census = df_census.drop(['education'], axis = 1)
df_census = pd.get_dummies(df_census)
df_census.head()
```

```python
df_census = df_census.drop(['income_ <=50K'], axis = 1)
df_census
```

```python
X = df_census.iloc[:, :-1]
y = df_census.iloc[:, -1]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
def cross_val(classifier, num_splits = 10):
  scores = cross_val_score(classifier, X, y , cv = num_splits)
  print('정확도: ', np.round(scores, 2))
  print('평균 정확도: ', np.round(np.mean(scores), 2))
```

```python
cross_val(LogisticRegression())
```

```python
from xgboost import XGBClassifier
```

```python
cross_val(XGBClassifier())
```

```python
cross_val(XGBClassifier(n_estimators = 5))
```
